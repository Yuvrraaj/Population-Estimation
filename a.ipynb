```python
import cv2
import matplotlib.pyplot as plt
import os

def draw_boxes(image_path, label_path):
    image = cv2.imread(image_path)
    if image is None:
        print(f"Error loading image: {image_path}")
        return

    height, width = image.shape[:2]

    if not os.path.exists(label_path):
        print(f"Label file not found: {label_path}")
        return

    with open(label_path, 'r') as f:
        for line in f:
            cls, x, y, w, h = map(float, line.strip().split())
            # Convert YOLO normalized coords to pixel coords
            x1 = int((x - w / 2) * width)
            y1 = int((y - h / 2) * height)
            x2 = int((x + w / 2) * width)
            y2 = int((y + h / 2) * height)
            cv2.rectangle(image, (x1, y1), (x2, y2), (255, 0, 0), 2)  # Blue boxes

    plt.figure(figsize=(8, 8))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()
```

```python
import os
import glob
import pandas as pd

# Path where YOLO prediction label files are stored
label_dir = '/yolo_predictions/predict/labels'  # Update path if needed

# Define class IDs
RESIDENTIAL_CLASS_ID = 0
NON_RESIDENTIAL_CLASS_ID = 1

# Function to count boxes in each class
def count_building_classes(label_path):
    res_count = 0
    nonres_count = 0

    with open(label_path, 'r') as f:
        for line in f:
            class_id = int(line.strip().split()[0])
            if class_id == RESIDENTIAL_CLASS_ID:
                res_count += 1
            elif class_id == NON_RESIDENTIAL_CLASS_ID:
                nonres_count += 1

    return res_count, nonres_count

# Store counts
counts = []

for label_file in sorted(glob.glob(os.path.join(label_dir, '*.txt'))):
    image_name = os.path.basename(label_file).replace('.txt', '.png')
    res_count, nonres_count = count_building_classes(label_file)
    total = res_count + nonres_count
    counts.append({
        'Image': image_name,
        'Residential_Building_Count': res_count,
        'Non_Residential_Building_Count': nonres_count,
        'Total_Detections': total
    })
    print(f"{image_name}: 🟪 {res_count} Residential | 🟧 {nonres_count} Non-Residential | Total: {total}")

# Save to CSV
df = pd.DataFrame(counts)
df.to_csv('residential_and_nonresidential_counts.csv', index=False)
print("\n✅ Saved to 'residential_and_nonresidential_counts.csv'")
```

```python
import cv2
import os

# Paths
image_dir = 'images'
label_dir = 'yolo_predictions/predict/labels'
output_dir = 'output2/visualized'

os.makedirs(output_dir, exist_ok=True)

# Colors
colors = {
    0: (255, 0, 255),  # Purple for residential
    1: (0, 165, 255)   # Orange for non-residential
}

# Class names (optional)
class_names = {
    0: 'Residential',
    1: 'Non-Residential'
}

def draw_yolo_boxes(image_name, include_non_residential=False):
    image_path = os.path.join(image_dir, image_name)
    label_path = os.path.join(label_dir, image_name.replace('.png', '.txt'))

    # Read image
    image = cv2.imread(image_path)
    if image is None:
        print(f"Image not found: {image_path}")
        return

    height, width, _ = image.shape

    # Read label file
    if not os.path.exists(label_path):
        print(f"Label file not found: {label_path}")
        return

    with open(label_path, 'r') as f:
        for line in f:
            parts = line.strip().split()
            class_id = int(parts[0])
            x_center, y_center, w, h = map(float, parts[1:])

            if class_id == 1 and not include_non_residential:
                continue  # Skip non-residential buildings if excluded

            # Convert normalized coords to pixel coords
            x1 = int((x_center - w / 2) * width)
            y1 = int((y_center - h / 2) * height)
            x2 = int((x_center + w / 2) * width)
            y2 = int((y_center + h / 2) * height)

            # Draw bounding box
            cv2.rectangle(image, (x1, y1), (x2, y2), colors[class_id], 2)
            cv2.putText(image, class_names[class_id], (x1, y1 - 5),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.5, colors[class_id], 2)

    # Save and show image
    output_path = os.path.join(output_dir, f"boxed_{image_name}")
    cv2.imwrite(output_path, image)
    print(f"✅ Output saved to: {output_path}")

# Example usage:
draw_yolo_boxes('tile_000_0051.png')  # Only residential boxes
# draw_yolo_boxes('tile_000_0052.png', include_non_residential=True)  # All boxes
```
